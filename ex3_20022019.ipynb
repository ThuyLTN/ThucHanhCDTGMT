```
from mtcnn.mtcnn import MTCNN
import cv2

img = cv2.imread("ivan.jpg")
detector = MTCNN()
detect_bound = detector.detect_faces(img)
```

```
print(detect_bound)
```

```
keypoints = detect_bound[0]['keypoints']
```

```
x=[]
y=[]
for item in keypoints.values():
  x.append(item[0])
  y.append(item[1])
print(x,y)
```

```
img.shape
```

```
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
%matplotlib inline

im = np.array(img, dtype=np.uint8)

# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(im)

# Create a Rectangle patch
rect = patches.Rectangle((83,93),169,216,linewidth=1,edgecolor='r',facecolor='none')

scatters = plt.scatter(x,y,c='red')

# Add the patch to the Axes
ax.add_patch(rect)
plt.axis('off')

plt.show()
```
